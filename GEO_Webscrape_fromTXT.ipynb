```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Path to your file containing individual codes
file_path = '/Users/noahnicol/Desktop/Desktop/School/2024-25/Lu Lab/aging_sample_info.txt'
```

```python

# Read the sample codes from the text file
with open(file_path, 'r') as f:
    sample_ids = [line.strip().split()[0] for line in f.readlines()]

# Base URL for accessing GEO pages
base_url = 'https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc='

# List to store the extracted data
data = []

# Function to extract information from each individual's page
def extract_info(soup, sample_id):
    info = {
        'ID': sample_id, 'Age': None, 'Gender': None,
        'Ethnicity': None, 'Tissue': None, 'Smoking': None
    }

    # Locate the 'Characteristics' section
    characteristics_section = soup.find('td', text='Characteristics')

    if characteristics_section:
        # Extract the next <td> containing demographic details
        details_td = characteristics_section.find_next_sibling('td')
        
        # Split the content by <br/> and iterate over the lines
        details = [line.strip() for line in details_td.get_text(separator="|").split('|')]

        # Loop through each detail and extract relevant info
        for detail in details:
            detail = detail.lower()

            if 'age' in detail:
                info['Age'] = detail.split(':')[-1].strip()
            elif 'gender' in detail:
                info['Gender'] = detail.split(':')[-1].strip()
            elif 'ethnicity' in detail:
                info['Ethnicity'] = detail.split(':')[-1].strip()
            elif 'tissue' in detail:
                info['Tissue'] = detail.split(':')[-1].strip()
            elif 'smoking' in detail:
                info['Smoking'] = detail.split(':')[-1].strip()

    return info

# Loop through each ID and extract data
for sample_id in sample_ids:
    url = base_url + sample_id
    print(f'Scraping {url}...')
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')
        try:
            individual_info = extract_info(soup, sample_id)
            data.append(individual_info)
        except Exception as e:
            print(f'Error extracting data for {sample_id}: {e}')
    else:
        print(f'Failed to retrieve {sample_id} (status code: {response.status_code})')
```

```python
# Create a DataFrame to store the scraped data
df = pd.DataFrame(data)

# Save the data to a CSV file
output_path = '/Users/noahnicol/Desktop/Desktop/School/2024-25/Lu Lab/geo_data.csv'
df.to_csv(output_path, index=False)

print(f'Scraping complete! Data saved to {output_path}')
```
